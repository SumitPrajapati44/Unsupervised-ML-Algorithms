```python
x = df.iloc[:,0:15]
x.shape
```

```python
y = df.iloc[:,-1]
y.unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier

knn = DecisionTreeClassifier(criterion='entropy', random_state=42)
```

```python
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

result = confusion_matrix(y_test,pred)
print("confusion_matrix...")
print(result)
print("Accuracy_score....\n",accuracy_score(y_test,pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

boost = AdaBoostClassifier(DecisionTreeClassifier(criterion='entropy', random_state=42))
boost.fit(x_train,y_train)
```

```python
y_b = boost.predict(x_test)
y_b
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

result = confusion_matrix(y_test,y_b)
print("confusion_matrix...")
print(result)
print("Accuracy_score....\n",accuracy_score(y_test,y_b))
```
